```python
import os
import face_recognition

def train_face_recognition(root_dataset_dir):
    known_encodings = {}

    for person_dir in os.listdir(root_dataset_dir):
        person_path = os.path.join(root_dataset_dir, person_dir)

        if not os.path.isdir(person_path):
            continue

        known_encodings[person_dir] = []

        for filename in os.listdir(person_path):
            image_path = os.path.join(person_path, filename)

            image = face_recognition.load_image_file(image_path)
            face_encodings = face_recognition.face_encodings(image)

            if len(face_encodings) == 1:
                known_encodings[person_dir].append(face_encodings[0])
            else:
                print(f"Skipping {filename} in {person_dir} as it doesn't have exactly one face.")

    return known_encodings

# Usage:
dataset_directory = "students"
known_encodings = train_face_recognition(dataset_directory)
```

```python
import cv2
import pandas as pd
from datetime import datetime
```

```python
def recognize_faces(known_encodings):
    video_capture = cv2.VideoCapture(0)

    while True:
        ret, frame = video_capture.read()

        if not ret:
            break

        face_locations = face_recognition.face_locations(frame)
        face_encodings = face_recognition.face_encodings(frame, face_locations)

        for face_encoding in face_encodings:
            # Default name if face is not recognized
            name = "Unknown"

            # Compare the current face encoding with known encodings
            for known_name, known_encodings_list in known_encodings.items():
                for known_encoding in known_encodings_list:
                    matches = face_recognition.compare_faces([known_encoding], face_encoding)

                    if True in matches:
                        name = known_name
                        break

                if name != "Unknown":
                    break

            # Print the recognized name
            print(f"Detected: {name}")

        cv2.imshow("Video", frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

    video_capture.release()
    cv2.destroyAllWindows()
```

```python
# # Usage:
# known_encodings_file = "known_encodings.pkl"
# known_encodings = load_known_encodings(known_encodings_file)

recognize_faces(known_encodings)
```
